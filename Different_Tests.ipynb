```python
print('Ingrese el número que desea evaluar si es un número primo:')
nro = input() 
print('\n')
nro = int(nro) #Casteo del input a evaluar

def verificación_primo(nro):
  es_primo = True

  for i in range(2,nro):
    print('i = ', i)
    if nro % i == 0:
      es_primo = False
      print('El número', nro, 'testeado NO ES UN NÚMERO PRIMO y es divisible por', i)
  if (es_primo == True):
      print('el número', nro, 'testeado, es un número PRIMO') #excepción para cuando el nro es primo
  return
  
print(verificación_primo(nro))
  
```

```python
print('Ingrese el número natural para aplicarle la función FACTORIAL (n!):')
numero = int(input())

def factorial (numero):
  if(numero>1):
   numero = numero * factorial(numero -1)
  return numero

print('Factorial', numero, '! es = ', factorial(numero))
```
